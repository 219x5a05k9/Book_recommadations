# USER RATING DATASET PREAPARATION

```python
ratings=pd.read_csv('/content/drive/MyDrive/Machine Learning stuff/Book_recommandations/BX-Book-Ratings.csv',sep=";",error_bad_lines=False,encoding='latin-1')
```

```python
ratings
```

```python
ratings.rename(columns={
    'User-ID':'user_id',
    'Book-Rating':'book_rating'},inplace=True
)
```

```python
# Merge the books and ratings 
ratings_books=ratings.merge(books,on='ISBN')
```

```python
ratings_books
```

```python
ratings_books.shape
```

```python
#Group by Operatins
num_rating=ratings_books.groupby('title')['book_rating'].count().reset_index()
```

```python
num_rating.tail(10)
```

```python
num_rating.rename(columns={
    "book_rating":"num_rating"
},inplace=True)
```

```python
num_rating.head(2)
```

```python
final_rating=ratings_books.merge(num_rating,on='title')
```

```python
final_rating.head(2)
```

```python
final_rating=final_rating[final_rating['num_rating']>50]
```

```python

final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
#This is for know the how many people have like same book and based on the their rating cluster them in way to make easier recommadation
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='book_rating')
```

```python
book_pivot.head()
```

```python
# whenever the occuring the nan values with fillna(0)
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
# Nearst neighbors algo well suitable for clustering
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')# brute for clusting algo
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=10)
#where n_neighbor = 6 means number of suggestion u want
```

```python
distance
```

```python
for i in range(len(suggestion)):
  print(book_pivot.index[suggestion[i]])
```

# Cross Checking

```python
def recommad(book_name):
  book_id=np.where(book_pivot.index==book_name)[0][0]
  distance,suggestion=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=10)
  for i in range(len(suggestion)):
    books=book_pivot.index[suggestion[i]]
    for j in books:
      print(j)
```

```python
book_name='2010: Odyssey Two'
recommad(book_name)
```
